Find parameters for the bionomial distribution such that the mean and variance are the same as in the normal distribution above (for \\( n > 0 \wedge \mu > 0 \wedge \sigma^{2} > \mu\\)).

\\begin{align}
\mu &= n p & \wedge \quad \sigma^{2} =& n p (1 - p) \\\\
p &= \frac{\mu}{n} & \wedge \quad \sigma^{2} =& \mu (1 - \frac{\mu}{n}) \\\\
 & & \sigma^{2} n =& \mu (n - \mu) \\\\
 & & (\sigma^{2} - \mu) n =& \mu^{2} \\\\
p &= \frac{\sigma^{2} - \mu}{\mu} & \wedge \quad n =& \frac{\mu^{2}}{\sigma^{2} - \mu}
\\end{align}

## Scaled Random Variables

Each element is scaled by an increasing factor.

\\begin{align}
E[a X] =& a E [X] = a \mu \\\\
Var[a X] =& a^{2} Var[X] = a^{2} \sigma^{2}
\\end{align}

```python
def scaling_iid(distribution, shape):
    events = distribution(shape)
    for i in xrange(shape[0]):
        factor = i + 1
        events[i, :] *= factor
    return (events.mean(axis=1), events.std(axis=1, ddof=1))
```

```python
data_scaling = list()
label_scaling = list()
for (dist, name) in izip(continuous_distributions, continuous_names):
    data_scaling.append(scaling_iid(dist, shape))
    label_scaling.append(name)
for (dist, name) in izip(discrete_distributions, discrete_names):
    data_scaling.append(scaling_iid(dist, shape))
    label_scaling.append(name)
```

## Sum of Multiple _iid_ Random Variables.

Each element is the sum of an increasing number of *iid* random variables.

\\begin{align}
E[a X + b Y] =& a E [X] + b E[Y] = (a + b) \mu \\\\
Var[a X + b Y] =& a^{2} Var[X] + b^{2} Var[Y] = (a^{2} + b^{2}) \sigma^{2}
\\end{align}

```python
def sum_iid(distribution, shape):
    events = distribution(shape)
    (num, t) = shape
    for i in xrange(num):
        events[i:, :] += dist((num - i, t))
    return (events.mean(axis=1), events.std(axis=1, ddof=1))
```

```python
data_sum = list()
label_sum = list()
for (dist, name) in izip(continuous_distributions, continuous_names):
    data_sum.append(sum_iid(dist, shape))
    label_sum.append(name)
for (dist, name) in izip(discrete_distributions, discrete_names):
    data_sum.append(sum_iid(dist, shape))
    label_sum.append(name)
```

## Sum of Multiple, Variably Scaled, _iid_ Random Variables

```python
def sum_var_iid(distribution, shape):
    events = distribution(shape)
    (num, t) = shape
    for i in xrange(num):
        factor = numpy.random.randint(1, 10 * num)
        events[i:, :] += factor * dist((num - i, t))
    return (events.mean(axis=1), events.std(axis=1, ddof=1))
```

```python
data_sum_var = list()
label_sum_var = list()
for (dist, name) in izip(continuous_distributions, continuous_names):
    data_sum_var.append(sum_var_iid(dist, shape))
    label_sum_var.append(name)
for (dist, name) in izip(discrete_distributions, discrete_names):
    data_sum_var.append(sum_var_iid(dist, shape))
    label_sum_var.append(name)
```

## Auto-regressive Process (induces correlations)

The notation \\( \text{AR}(p) \\) indicates an autoregressive model of order \\( p \\). The \\( \text{AR}(p) \\) model is defined as

\\begin{align}
X\_{i} &= c\_{i} + \sum^{p}\_{j = 1} \varphi\_{j} X\_{i - j} + \epsilon\_{i} \\\\
X\_{0} &= c\_{0} + \epsilon\_{0}, \text{where} \\\\
i &\in [0, N - 1] \\\\
c\_{i} & \text{is some individual constant} \\\\
\varphi\_{j} & \text{denotes the correlation strength between element } i \text{ and the } j \text{-th previous element}
\\end{align}

\\begin{align}
E [X\_{0}] =& c\_{0} + \mu\_{\epsilon\_{0}} \\\\
Var [X\_{0}] =& \sigma^{2}\_{\epsilon\_{0}} \\\\
E [X\_{i}] =& c\_{i} + \sum^{p}\_{j = 1} \varphi\_{j} E[X\_{i - j}] + \mu\_{\epsilon\_{i}} \\\\
Var [X\_{i}] =& Var[\sum^{p}\_{j = 1} \varphi\_{j} X\_{i - j}] + \sigma^{2}\_{\epsilon\_{i}} = \sum^{p}\_{j = 1} \varphi\_{j}^{2} Var[X\_{i - j}] + \sigma^{2}\_{\epsilon\_{i}} \\\\
Std[X\_{i}] =& \left( \sum^{p}\_{j = 1} \varphi\_{j}^{2} Var[X\_{i - j}] + \sigma^{2}\_{\epsilon\_{i}} \right)^{\frac{1}{2}}
\\end{align}

Assuming that \\( \epsilon\_{i} \\) is the same for all elements \\( i \\), i.e.,
\\begin{align}
E[\epsilon\_{i}] =& \mu\_{\epsilon} \quad \forall i \in [0, N - 1] \\\\
Var[\epsilon\_{i}] =& \sigma^{2}\_{\epsilon} \quad \forall i \in [0, N - 1] \\\\
\\end{align}
then \\( Var[X\_{i - j}] \\) is some factorisation \\( k \\) of \\( \sigma^{2}\_{\epsilon} \\) such that
\\begin{align}
Var[X\_{i - j}] =& k\_{i - j} \sigma^{2}\_{\epsilon} \\\\
Std[X\_{i}] =& \left( \sum^{p}\_{j = 1} \varphi\_{j}^{2} k\_{i - j} \sigma^{2}\_{\epsilon} + \sigma^{2}\_{\epsilon} \right)^{\frac{1}{2}}
= \left( \left[ \sum^{p}\_{j = 1} \varphi\_{j}^{2} k\_{i - j} + 1 \right] \sigma^{2}\_{\epsilon} \right)^{\frac{1}{2}}
= \sigma\_{\epsilon} \left( \sum^{p}\_{j = 1} \varphi\_{j}^{2} k\_{i - j} + 1 \right)^{\frac{1}{2}}
\\end{align}

For extreme case \\( \varphi\_{j} = 0 \\):
\\begin{align}
Std[X\_{i}] =& \sigma\_{\epsilon} \left( \sum^{p}\_{j = 1} \varphi\_{j}^{2} k\_{i - j} + 1 \right)^{\frac{1}{2}}
\\end{align}

Working on the general case... AR(2)

\\begin{align}
Var[X\_{0}] =& \sigma^{2}\_{\epsilon} \\\\
Var[X\_{1}] =& \varphi^{2}\_{1} Var[X\_{0}] + \sigma^{2}\_{\epsilon} = \left( \varphi^{2}\_{1} + 1 \right) \sigma^{2}\_{\epsilon} \\\\
Var[X\_{2}] =& \varphi^{2}\_{1} Var[X\_{1}] + \varphi^{2}\_{2} Var[X\_{0}] + \sigma^{2}\_{\epsilon}
= \left( \varphi^{2}\_{1} + 1 \right) \varphi^{2}\_{1} \sigma^{2}\_{\epsilon} + \varphi^{2}\_{2} \sigma^{2}\_{\epsilon} + \sigma^{2}\_{\epsilon}
= \left( \left( \varphi^{2}\_{1} + 1 \right) \varphi^{2}\_{1} + \varphi^{2}\_{2} + 1 \right) \sigma^{2}\_{\epsilon} \\\\
\\end{align}

```python
def auto_regressive(order, constant, phi, noise, num_elements, time):
    """
    the factors in phi must be in reverse order.
    
    Say the correlation strength with the j-th element is in decreasing order:
        \phi_{1} > \phi{2} > \phi{3}
    then the array phi should be ordered in reverse:
        phi[0] < phi[1] < phi[2]
    """
    events = numpy.zeros((num_elements, time))
    for t in xrange(time):
        events[0, t] = constant[0] + noise()
        for p in xrange(1, order + 1):
            events[p, t] = constant[p] + (phi[:p] * events[:p, t]).sum() + noise()
        for i in xrange(order + 1, num_elements):
            j = i - order
            events[i, t] = constant[i] + (phi * events[j:i, t]).sum() + noise()
    return (events.mean(axis=1), events.std(axis=1, ddof=1))
```

```python
def write_to_csv(file_h, mean, std, order, phi):
    for info in izip(mean, std, repeat(order), repeat(phi)):
        file_h.write("%f,%f,%d,%f\n" % info)
```

```python
small_elements = 100
# constant for each element
constant = numpy.zeros(small_elements)
constant.fill(1.0)
#constant = numpy.arange(small_elements)
# noise term epsilon
intensity = 0.2
noise = lambda : intensity * numpy.random.standard_normal()
#
data_ar = list()
label_ar = list()
#file_h = open("auto_regressive.csv", "w")
#file_h.write("mean,std,order,phi\n")
for p in [2, 3, 4, 5]:
    # correlation strength
    phi = numpy.zeros(p)
#    for strength in numpy.linspace(0.3, 0.6, 5):
    for strength in [1.0]:
        phi.fill(strength)
        (mean, std) = auto_regressive(p, constant, phi, noise, small_elements, time)
#        write_to_csv(file_h, mean, std, p, strength)
        data_ar.append((mean, std))
        label_ar.append("AR$(%d), \\varphi = %.3f$" % (p, strength))
#file_h.close()
```
